```python
from bs4 import BeautifulSoup
import requests


pg=requests.get("https://www.flipkart.com/search?as=off&as-show=on&count=40&otracker=start&p%5B%5D=facets.price_range.from%3D1500&p%5B%5D=facets.price_range.to%3DMax&p%5B%5D=sort%3Dpopularity&q=casuals+shoe&sid=osp%2Fcil%2Fe1f")
script=BeautifulSoup(pg.content,'html.parser')
print(script.prettify())
```

```python
pn=script.find_all('a',class_='_2cLu-l')
ps=script.find_all('div',class_='hGSR34')
pr=script.find_all('div',class_='_1vC4OE')
```
